```python
import pandas as pd 
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# import data
coffee = pd.read_csv("../data/crop/coffee_output.csv")
```

```python
# TODO: view first 5 rows of your data
coffee.head()
```

```python
# TODO: describe your numeric columns
coffee.describe()
```

```python
# TODO: display the null values in your dataframe
coffee.isna().sum()
```

```python
# TODO: drop your null values
coffee.dropna(inplace=True)
```

```python
# TODO: Select only Minas Gerais
MG_coffee = coffee[coffee["subdivision"] == "Minas Gerais"]
```

```python
# TODO: BEGIN YOUR EDA!
# Display a line plot of bearing trees changes over the years
sns.lineplot(MG_coffee, x = MG_coffee["year"], y = MG_coffee["bearing_trees"])
```

```python
# Display a line plot of bearing hectares changes over the years
sns.lineplot(MG_coffee, x = MG_coffee["year"], y = MG_coffee["bearing_hectares"])
```

```python
# Display a line plot of the count of million 60kgs bags of coffee changes over the years
sns.lineplot(MG_coffee, x = MG_coffee["year"], y = MG_coffee["60kgs_bag"])
```

```python
# Calculate the correlation coefficient between the count of 60kgs bag of coffee and bearing trees
# using corr() method
print(MG_coffee[["60kgs_bag","bearing_trees"]].corr())
# Display the scatter plot of bearing_trees and 60kgs_bag
sns.scatterplot(MG_coffee,x="bearing_trees",y="60kgs_bag")
```

```python
# Calculate the correlation coefficient between the count of 60kgs bag of coffee and bearing hectares
# using corr() method
print(MG_coffee[["60kgs_bag","bearing_hectares"]].corr())
# Display the scatter plot of bearing_hectares and 60kgs_bag
sns.scatterplot(MG_coffee,x="bearing_hectares",y="60kgs_bag")
```
